```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def linear_func(a,x,b):
  return a*x + b
```

```python
def non_linear_func(a,x,b):
  return a*x*x + b
```

```python
def sin_func(a,x,b):
  return a*(np.sin(x)) + b
```

```python
def exp_func(a,x,b):
  return a*np.exp(x) + b
```

NO-01 : y = ax+b where x = [-10,10]. Now I have considered a = 5 and b = 1


```python
x = np.random.uniform(-10,10,50)
# x = np.arange(-10,10,1)
y = linear_func(5,x,1)
```

```python
print(x)
```

```python
plt.plot(x,y, color='red' ,marker = 'o',label= 'y=5x+1')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-10, 10)
plt.ylim(-30, 30)
plt.title(f"Straight line for y = 5x+1")
plt.legend()
# plt.axis('off')
plt.show()
```

No-02 : y = aSin(x) + b where x = [-10,10]

```python
x = np.arange(-10,10,0.01)

y = sin_func(5,x,1)
```

```python
plt.plot(x,y, color='blue' ,linewidth = 5,label= 'y=5sinx+1')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-10, 10)
plt.ylim(-20, 20)
plt.title(f"Sin function for y = 5sinx + 1")
plt.legend()
plt.grid(True)
plt.show()
```

No -03: y = ax^2 + b where x=[-10,10]


```python
x = np.arange(-10,10,0.5)
y = non_linear_func(1,x,5)
```

```python
plt.plot(x,y, color='green' ,marker = 'o',linewidth=2,label= 'y=x^2+5')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-10, 10)
plt.ylim(-20, 30)
plt.title(f"non linear function for y = x^2 + 5")
plt.legend()
plt.grid(True)
plt.show()
```

No-04: y = ae^x + b where range of x = [-10,10]

```python
x = np.arange(-10,10,0.5)
y = exp_func(3,x,1)
```

```python
plt.plot(x,y, color='red' ,marker = 'o',linewidth=2,label= 'y=3e^x + 1')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-10, 10)
plt.ylim(-20, 50)
plt.title(f"exponential function for y = 3e^x + 1")
plt.legend()
plt.grid(True)
plt.show()
```

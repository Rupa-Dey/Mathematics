```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def non_linear_sinfunc(a,x,b,f):
  return a*(np.sin(2*np.pi*f*x)) + b
```

```python
def non_linear_cosfunc(a,x,b,f):
  return a*(np.cos(2*np.pi*f*x)) + b
```

```python
x = np.arange(0,10,0.001)

y1 = non_linear_sinfunc(2,x,1,1)
y2 = non_linear_sinfunc(1.3,x,-2,3)
y3 = non_linear_cosfunc(0.5,x,2,1.5)
y = y1+y2+y3
```

```python
plt.figure(figsize=(12,8))
plt.plot(x,y1,color='blue', label='y1')
plt.plot(x,y2,color='green', label='y2')
plt.plot(x,y3,color='red', label='y3')
plt.plot(x,y, color='orange', label='y')
plt.title("composite Signal")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.legend()
plt.show()
```
